```python
#Câu 1
from math import sqrt
print("Chương trình tính diện tích Tam Giác")
a=float(input("Nhập cạnh a>0:"))
b=float(input("Nhập cạnh b>0:"))
c=float(input("Nhập cạnh c>0:"))
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a:
    print("Tam giác không hợp lệ")
else:
    cv=a+b+c
    p=cv/2
    dt=sqrt(p*(p-a)*(p-b)*(p-c))
    print("Diện tích =",dt)
```

```python
#Câu 2
from random import randrange
while True:
    somay=randrange(1,101)
    solandoan=0
    win=False
    while solandoan<7:
        solandoan+=1
        songuoi=int(input("Máy đoán [1..100], mời bạn đoán:"))
        print("Bạn đoán lần thứ ",solandoan)
        if somay==songuoi:
            print("Chúc mừng bạn đoán đúng, số máy là=",somay)
            win=True
            break
        if somay>songuoi:
            print("Bạn đoán sai, số máy > số bạn")
        elif somay<songuoi:
            print("Bạn đoán sai, số máy < số bạn")
    if win==False:
        print("GAME OVER!, số máy =",somay)
    hoi=input("Tiếp không? (k/c)")
    if hoi=="k" or hoi=="K":
        break
    elif hoi =="c" or hoi=="C":
        continue
print("Cám ơn bạn đã chơi Game!")
```

```python
#Câu 3
def BMI(weight,height):
    return weight/(height*height)
def PhanLoai(bmi):
    if bmi<18.5:
        return "Gầy"
    elif bmi<=24.9:
        return "Bình thường"
    elif bmi<=29.9:
        return "Hơi béo"
    elif bmi<=34.9:
        return "Béo phì cấp độ I"
    elif bmi<=39.9:
        return "Béo phì cấp độ II"
    else:
        return "Béo phì cấp độ III"
def NguyCoBenh(bmi):
    if bmi<18.5:
        return "Nguy cơ bệnh thấp"
    elif bmi<=24.9:
        return "Nguy cơ bệnh trung bình"
    elif bmi<=29.9:
        return "Nguy cơ bệnh cao"
    elif bmi<=34.9:
        return "Nguy cơ bệnh cao"
    elif bmi<=39.9:
        return "Nguy cơ bệnh rất cao"
    else:
        return "Nguy cơ bệnh nguy hiểm"
print("---Chương trình tính chỉ số BMI---")
print("Nhập chiều cao (m):")    
height = float(input())
print("Nhập cân nặng (kg):")
weight = float(input())
bmi = BMI(weight, height)
print("Chỉ số BMI của bạn là:", bmi)
print("Phân loại:", PhanLoai(bmi))
print("Nguy cơ bệnh:", NguyCoBenh(bmi))
```

```python
#Câu 4
def ROI(dt,cp):
    return (dt-cp)/cp
def GoiYDauTu(roi):
    if roi>=0.75:
        return "Nên đầu tư "
    else:
        return "Không nên đầu tư"
print("---Chương trình tính ROI---")
dt=int(input("Nhập doanh thu :"))
cp=int(input("Nhập chi phí :"))
roi=ROI(dt,cp)
print("Tỉ lệ ROI =",roi)
print("==>>",GoiYDauTu(roi))
```

```python
def Fibonacci(n):
    if n<=2:
        return 1
    return Fibonacci(n-1)+Fibonacci(n-2)
def Listfibo(n):
    for i in range(1,n+1):
        print(Fibonacci(i),end="\t")
print("--Nhập vị trí cần tính Fibonacci:")
n=int(input())
print("Số Fibonacci tại vị trí",n,"là: ",Fibonacci(n))
print("\nDãy Fibonacci đến vị trí",n,":")
Listfibo(n)
```

```python
import math
def DoDai(x1,x2,y1,y2):
    return math.sqrt((x2-x1)**2+(y2-y1)**2)
print("Nhập tọa độ điểm A (x1,y1):")
x1=float(input("x1="))
y1=float(input("y1="))
print("Nhập tọa độ điểm B (x2,y2):")
x2=float(input("x2="))
y2=float(input("y2="))
print("Độ dài giữa A và B là:",round(DoDai(x1 ,x2,y1,y2),2))
```

```python
#Câu 9: căn 2, n dấu căn lồng nhau
def Can2(n):
    if n==1:
        return math.sqrt(2)
    return math.sqrt(2+Can2(n-1))
n=int(input("Nhập n>0:"))
if n<=0:
    print("n phải >0, mời nhập lại")
    n=int(input("Nhập n>0:"))
print("Kết quả là:",round(Can2(n),2))
```

```python
#Câu 10: Vẽ hình
import time
import os

def clear_screen():
    os.system('cls' if os.name == 'nt' else 'clear')
n=int(input("Nhập chiều cao của hình = "))
#Hình 1
def hinh1():
    base_indent=2*n
    for i in range(1,n+1):
        print(" "*base_indent+"* "*i)
    print(" ".join("*"for _ in range(2*n+1)))
    for i in range(n,0,-1):
        print("* "*i)

#Hình 2
def hinh2():
    base_indent=2*n
    for i in range(1,n+1):
        if i==1:
            print(" "*base_indent+"*")
        else:
            print(" "*base_indent+"*"+" "*(2*i-3)+"*")
    print(" ".join("*"  for _ in range(2*n+1)))
    for i in range(n,0,-1):
        if i==1:
            print("*")
        else:
            print("*"+" "*(2*i-3)+"*")
#Hình 3
def hinh3():
    for i in range(n,0,-1):
        print(" "*(n-1)+"*"*i)
    for i in range(2,n+1):
        print(" "*(n-i)+"*"*i)
#Hình 4
def hinh4():
    for i in range(n,0,-1):
        if i==1:
            print(" "*(n-1)+"*")
        elif i==n:
            print(" "*(n-1)+"*"*i)
        else:
            print(" "*(n-1)+"*"+" "*(i-2)+"*")
    for i in range(2,n+1):
        if i==1:
            print("*"*i)
        else:
            print(" "*(n-i)+"*"+" "*(i-2)+"*")
#
shapes=[hinh1,hinh2,hinh3,hinh4]
for draw in shapes:
    clear_screen()
    draw()
    time.sleep(5)
```

```python
#Câu 11
def sum1(n):
    s = 0
    while n > 0:
        s += 1
        n -= 1
    return s
def sum2():
    global val
    s = 0
    while val > 0:
        s += 1
        val -= 1
    return s
def sum3():
    s = 0
    for i in range(val, 0, -1):
        s += 1
    return s
#Trường hợp 1:5 5 0
def main():
    global val
    val = 5
    print("Trường hợp 1:")
    print(sum1(5))
    print(sum2())
    print(sum3())
main()

#Trường hợp 2: 5 5 5
def main():
    global val
    val = 5
    print("Trường hợp 2:")
    print(sum1(5))
    print(sum3())
    print(sum2())
main()
#Trường hợp 3:5 5 0
def main():
    global val
    val = 5
    print("Trường hợp 3:")
    print(sum2())
    print(sum1(5))
    print(sum3())
main()
```

```python
#Câu 12: Hàm oscillate
def oscillate(start,end):
    result=[]
    for i in range(start,end):
        if i<0:
            result.append(i)
            result.append(-i)
        elif i==0:
            result.append(0)
            result.append(0)
        else:
            result.append(i)
            result.append(-i)
    return result
for n in oscillate(-3,5):
    print(n,end=" ")
print()
```

```python
#Câu 13: Kt số hoàn thiện, số thịnh vượng
def TongUoc(n):
    s=0
    for i in range(1,n):
        if n%i==0:
            s+=i
    return s
def KtSoHoanThien(n):
    if TongUoc(n)==n:
        return True
    return False
def KtSoThinhVuong(n):
    if TongUoc(n)>n:
        return True
    return False
print("Nhập n>0:")
n=int(input())
if n<=0:
    print("n phải >0, mời nhập lại")
    n=int(input("Nhập n>0:"))
if KtSoHoanThien(n):
    print(n,"là số hoàn thiện")
else:
    print(n,"không là số hoàn thiện")
if KtSoThinhVuong(n):
    print(n,"là số thịnh vượng")
else:
    print(n,"không là số thịnh vượng")
```
